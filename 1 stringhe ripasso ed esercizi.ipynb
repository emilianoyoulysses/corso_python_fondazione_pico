### Creazione e operazioni base

```python
# Creazione stringhe
stringa1 = "Ciao"
stringa2 = 'Python'
stringa3 = """Stringa
multilinea"""

# Concatenazione
saluto = stringa1 + " " + stringa2
print(saluto)  # Output: Ciao Python

# Ripetizione
asterischi = "*" * 10
print(asterischi)  # Output: **********

### Indicizzazione e slicing

testo = "Python Programming"
# Indicizzazione (0-based)
print(testo[0])    # P
print(testo[-1])   # g

# Slicing [start:stop:step]
print(testo[0:6])   # Python
print(testo[7:])    # Programming
print(testo[::2])   # Pto rgamn

### Metodi principali

esempio = "  Hello World  "

# Modifica case
print(esempio.upper())       # "  HELLO WORLD  "
print(esempio.lower())       # "  hello world  "
print(esempio.title())       # "  Hello World  "

# Pulizia spazi
print(esempio.strip())       # "Hello World"
print(esempio.lstrip())      # "Hello World  "
print(esempio.rstrip())      # "  Hello World"

# Ricerca e sostituzione
print(esempio.find("World")) # 8
print(esempio.replace("World", "Python"))  # "  Hello Python  "

# Split e join
parole = "uno,due,tre".split(",")
print(parole)  # ['uno', 'due', 'tre']
print("-".join(parole))  # "uno-due-tre"

# Controlli
print("Hello".startswith("He"))  # True
print("World".endswith("ld"))    # True
print("123".isdigit())           # True
print("abc".isalpha())           # True
```

## SOLUZIONI

```python
### Soluzione Esercizio 1:
nome_completo = "Mario Rossi"
# Metodo 1: con split()
parti = nome_completo.split()
nome = parti[0]
cognome = parti[1]
print(f"Nome: {nome}, Cognome: {cognome}")

# Metodo 2: con find() e slicing
spazio = nome_completo.find(" ")
nome = nome_completo[:spazio]
cognome = nome_completo[spazio+1:]
print(f"Nome: {nome}, Cognome: {cognome}")

### Soluzione Esercizio 2:
prodotto = "Laptop"
prezzo = 899.99

# Metodo 1: f-string (Python 3.6+)
frase1 = f"Il prodotto {prodotto} costa {prezzo} euro"
print(frase1)

# Metodo 2: format()
frase2 = "Il prodotto {} costa {} euro".format(prodotto, prezzo)
print(frase2)

# Metodo 3: % formatting
frase3 = "Il prodotto %s costa %.2f euro" % (prodotto, prezzo)
print(frase3)

### Soluzione Esercizio 3:
def is_email_valid(email):
    # Controlla se contiene @
    if "@" not in email:
        return False
    
    # Divide in parte locale e dominio
    parti = email.split("@")
    if len(parti) != 2:
        return False
    
    # Controlla se c'è almeno un punto nel dominio
    dominio = parti[1]
    return "." in dominio

# Test
print(is_email_valid("test@email.com"))  # True
print(is_email_valid("test.email"))      # False

### Soluzione Esercizio 4:
testo = "Il gatto sale sul tetto. Il tetto è alto."

# Pulizia del testo e conversione in minuscolo
testo_pulito = testo.lower().replace(".", "")
parole = testo_pulito.split()

# Conteggio con dizionario
conteggio = {}
for parola in parole:
    if parola in conteggio:
        conteggio[parola] += 1
    else:
        conteggio[parola] = 1

print("Conteggio parole:")
for parola, count in conteggio.items():
    print(f"{parola}: {count}")
```
