# `.read_csv()`
Il metodo `read_csv()` viene utilizzato per leggere un file CSV (Comma-Separated Values) e convertirlo in un DataFrame, che puoi poi analizzare e manipolare in Python.

```python
mpg = pd.read_csv("https://raw.githubusercontent.com/emilianoyoulysses/corso_python_fondazione_pico/refs/heads/DataFrames-%26-TimeSeries/mpg.csv")
mpg
```

# `.count`
Il comando `.count()` restituisce il numero di valori non nulli (non NaN) per ciascuna colonna del DataFrame mpg.

```python
print(mpg.count())
```

# `.head()`
Il metodo `.head()` viene utilizzato per visualizzare le prime righe di un DataFrame o di una Series.

```python
mpg.head()
```

```python
mpg.tail(3) # visualizza gli ultimi elementi
```

```python
columns_of_interest = ["displ", "cty", "hwy"]
mpg[columns_of_interest].head()
```

```python
mpg[["cty", "hwy"]].head()
```

```python
mpg[["cty", "hwy"]].mean(axis=0) # valuto le colonna selezionate
```

```python
mpg[["cty", "hwy"]].head()
```

```python
mpg[["cty", "hwy"]].mean(axis=1).head() # su ogni riga ed i primi
```

```python
mpg["average_mileage"] = mpg[["cty", "hwy"]].mean(axis=1) # riga per riga
mpg.head()
```

```python
mpg.cyl.value_counts() # mostra i valori unici e la loro distribuzione
```

> Restituisce una Series booleana: `True` per le righe dove il valore nella colonna `manufacturer` è `"audi"`, `False` altrimenti. **Quante audi ho nel set di dati?**

```python
(mpg.manufacturer == "audi").sum()
```

```python
(mpg.cyl == 4).mean()
```

```python
mpg[mpg.cyl == 5] # uso Bool per filtrare il DF
```

> ### Possiamo mescolare ed abbinare le colonne utilizzate per generare booleani.

```python
mpg[(mpg.model == "maxima") | (mpg.cyl == 5)]
```

> ### AND limita le opzioni. Entrambi i predicati devono essere veri


```python
mpg[(mpg["class"] == "midsize") & (mpg["displ"] < 2)]
```

> ### Creiamo un nuovo DF

```python
original_df = pd.DataFrame({"x": [1, 2, 3]})
original_df
```

> ### Creiamo un alias, cioè un secondo nome che punta allo stesso oggetto in memoria.
> * `new_df` e `original_df` sono lo stesso oggetto.
> * Se modifichi uno, cambierà anche l'altro.

```python
new_df = original_df
```

```python
original_df["y"] = original_df.x * 100
original_df
```

```python
new_df
```

> ### Creo una vera copia indipendente del DataFrame `original_df`
> * `new_df` è una copia completa del contenuto di `original_df`
> * Ora possiamo modificare `new_df` senza toccare `original_df`

```python
new_df = original_df.copy()
new_df
```

```python
original_df["z"] = 5000
original_df
```

```python
new_df
```

> ### Ordino per singola colonna. 

```python
mpg.sort_values(by=["displ"], ascending=False, ignore_index=True).head()
```

> ### Ordino per più colonne. 

```python
mpg.sort_values(by=["hwy", "cty"], ascending=False, ignore_index=True).head()
```

> Aggiornamento del dataframe per memorizzare i valori ordinati \
> Utilizzo `inplace=True` o riassegna la variabile del dataframe (non entrambe)

```python
mpg.sort_values(by="displ", ascending=False, ignore_index=True, inplace=True)
mpg.head()
```
