## SOLUZIONI

```python
### Soluzione Esercizio 1:
def calcolatrice(a, b, operazione):
    if operazione == '+':
        return a + b
    elif operazione == '-':
        return a - b
    elif operazione == '*':
        return a * b
    elif operazione == '/':
        if b != 0:
            return a / b
        else:
            return "Errore: divisione per zero"
    else:
        return None

# Test
print(calcolatrice(10, 5, '+'))  # 15
print(calcolatrice(10, 5, '/'))  # 2.0

### Soluzione Esercizio 2:
def celsius_to_fahrenheit(celsius):
    # Formula: F = C * 9/5 + 32
    return celsius * 9/5 + 32

def fahrenheit_to_celsius(fahrenheit):
    # Formula: C = (F - 32) * 5/9
    return (fahrenheit - 32) * 5/9

# Test
print(celsius_to_fahrenheit(0))    # 32.0
print(fahrenheit_to_celsius(98.6)) # 37.0

### Soluzione Esercizio 3:
def statistiche(numeri):
    # Media
    media = sum(numeri) / len(numeri)
    
    # Mediana
    numeri_ordinati = sorted(numeri)
    n = len(numeri_ordinati)
    
    if n % 2 == 0:
        # Se pari, media dei due valori centrali
        mediana = (numeri_ordinati[n//2 - 1] + numeri_ordinati[n//2]) / 2
    else:
        # Se dispari, valore centrale
        mediana = numeri_ordinati[n//2]
    
    return {'media': media, 'mediana': mediana}

# Test
numeri = [5, 2, 8, 3, 9, 1, 7]
print(statistiche(numeri))  # {'media': 5.0, 'mediana': 5}

### Soluzione Esercizio 4:
def is_primo(n):
    if n < 2:
        return False
    
    # Controlla divisibilità fino alla radice quadrata di n
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    
    return True

# Test
print(is_primo(17))  # True
print(is_primo(20))  # False

# Bonus: primi numeri primi fino a 20
primi = [n for n in range(2, 21) if is_primo(n)]
print(f"Numeri primi fino a 20: {primi}")
```
