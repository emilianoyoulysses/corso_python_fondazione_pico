## PARTE 1: RIPASSO (3 minuti)

```python
### Valori booleani e creazione

# I due valori booleani
vero = True
falso = False

# Conversione a booleano con bool()
print(bool(1))      # True - numeri diversi da 0
print(bool(0))      # False - zero
print(bool("ciao")) # True - stringhe non vuote
print(bool(""))     # False - stringa vuota
print(bool([1,2]))  # True - liste non vuote
print(bool([]))     # False - lista vuota
print(bool(None))   # False - None è sempre False

### Operatori di confronto

a, b = 5, 3

# Confronti numerici
print(a > b)   # True - maggiore
print(a < b)   # False - minore
print(a >= b)  # True - maggiore o uguale
print(a <= b)  # False - minore o uguale
print(a == b)  # False - uguale
print(a != b)  # True - diverso

# Confronti con stringhe
print("python" == "Python")  # False - case sensitive
print("abc" < "abd")         # True - ordine lessicografico

### Operatori logici

x, y = True, False

# AND - True solo se entrambi True
print(x and y)      # False
print(True and True) # True

# OR - True se almeno uno è True
print(x or y)       # True
print(False or False) # False

# NOT - inverte il valore
print(not x)        # False
print(not y)        # True

# Precedenza: not > and > or
print(True or False and False)  # True (and valutato prima)
print((True or False) and False) # False (parentesi cambiano ordine)

### Operatori speciali

# is - identità oggetto
a = [1, 2, 3]
b = [1, 2, 3]
c = a
print(a == b)  # True - stesso contenuto
print(a is b)  # False - oggetti diversi
print(a is c)  # True - stesso oggetto

# in - appartenenza
print(3 in [1, 2, 3, 4])  # True
print("a" in "ciao")      # True
print(5 not in [1, 2, 3]) # True

### Short-circuit evaluation

# AND si ferma al primo False
def f1():
    print("f1 chiamata")
    return False

def f2():
    print("f2 chiamata")
    return True

risultato = f1() and f2()  # f2 non viene chiamata!

# OR si ferma al primo True
risultato = True or f2()   # f2 non viene chiamata!
```

## SOLUZIONI

```python
## Soluzione Esercizio 1:
def password_valida(password):
    # Controlla lunghezza
    if len(password) < 8:
        return False
    
    # Controlla presenza numero
    ha_numero = any(char.isdigit() for char in password)
    
    # Controlla presenza maiuscola
    ha_maiuscola = any(char.isupper() for char in password)
    
    return ha_numero and ha_maiuscola

# Versione alternativa più esplicita:
def password_valida_v2(password):
    if len(password) < 8:
        return False
    
    ha_numero = False
    ha_maiuscola = False
    
    for char in password:
        if char.isdigit():
            ha_numero = True
        if char.isupper():
            ha_maiuscola = True
    
    return ha_numero and ha_maiuscola

# Test
print(password_valida("Abc12345"))  # True
print(password_valida("abc12345"))  # False
print(password_valida("Abcdefgh"))  # False

### Soluzione Esercizio 2:
def is_bisestile(anno):
    # Metodo 1: seguendo la logica passo passo
    if anno % 400 == 0:
        return True
    if anno % 100 == 0:
        return False
    if anno % 4 == 0:
        return True
    return False

# Metodo 2: in una sola espressione
def is_bisestile_v2(anno):
    return (anno % 400 == 0) or (anno % 4 == 0 and anno % 100 != 0)

# Test
print(is_bisestile(2000))  # True
print(is_bisestile(1900))  # False
print(is_bisestile(2024))  # True

### Soluzione Esercizio 3:
def in_range_speciali(n):
    # Metodo 1: con or
    return (0 <= n <= 10) or (50 <= n <= 60) or (90 <= n <= 100)

# Metodo 2: con una lista di range
def in_range_speciali_v2(n):
    ranges = [(0, 10), (50, 60), (90, 100)]
    return any(start <= n <= end for start, end in ranges)

# Test
print(in_range_speciali(5))   # True
print(in_range_speciali(55))  # True
print(in_range_speciali(75))  # False

### Soluzione Esercizio 4:
def logica_speciale(a, b, c):
    # Conta quanti sono True
    count_true = sum([a, b, c])
    
    # Almeno due True
    if count_true >= 2:
        return True
    
    # Primo True e gli altri False
    if a and not b and not c:
        return True
    
    return False

# Versione compatta
def logica_speciale_v2(a, b, c):
    return sum([a, b, c]) >= 2 or (a and not b and not c)

# Test
print(logica_speciale(True, True, False))   # True
print(logica_speciale(True, False, False))  # True
print(logica_speciale(False, True, False))  # False
```
