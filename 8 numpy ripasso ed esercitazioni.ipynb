```python
### Soluzione Esercizio 1:
def analizza_matrice(matrice):
    risultati = {}
    
    # Medie
    risultati['media_righe'] = np.mean(matrice, axis=1)
    risultati['media_colonne'] = np.mean(matrice, axis=0)
    
    # Massimo
    max_val = np.max(matrice)
    max_pos = np.argmax(matrice)
    max_riga, max_col = np.unravel_index(max_pos, matrice.shape)
    risultati['max'] = {'valore': max_val, 'posizione': (max_riga, max_col)}
    
    # Minimo
    min_val = np.min(matrice)
    min_pos = np.argmin(matrice)
    min_riga, min_col = np.unravel_index(min_pos, matrice.shape)
    risultati['min'] = {'valore': min_val, 'posizione': (min_riga, min_col)}
    
    return risultati

# Test
mat = np.array([[1, 5, 3],
                [4, 2, 6],
                [7, 8, 9]])
print(analizza_matrice(mat))

### Soluzione Esercizio 2:
def normalizza_e_pulisci(arr, min_val=0, max_val=1, percentile_basso=5, percentile_alto=95):
    # Copia per non modificare originale
    arr_copia = arr.copy()
    
    # Calcola percentili e mediana
    p_basso = np.percentile(arr_copia, percentile_basso)
    p_alto = np.percentile(arr_copia, percentile_alto)
    mediana = np.median(arr_copia)
    
    # Sostituisci outlier con mediana
    arr_copia[arr_copia < p_basso] = mediana
    arr_copia[arr_copia > p_alto] = mediana
    
    # Normalizza
    arr_min = arr_copia.min()
    arr_max = arr_copia.max()
    
    if arr_max - arr_min > 0:
        arr_norm = (arr_copia - arr_min) / (arr_max - arr_min)
        arr_norm = arr_norm * (max_val - min_val) + min_val
    else:
        arr_norm = np.full_like(arr_copia, (min_val + max_val) / 2)
    
    return arr_norm

# Test
dati = np.random.normal(50, 15, 100)
dati[0] = 200  # Outlier
dati[1] = -50  # Outlier
puliti = normalizza_e_pulisci(dati)
print(f"Range originale: [{dati.min():.1f}, {dati.max():.1f}]")
print(f"Range pulito: [{puliti.min():.1f}, {puliti.max():.1f}]")

### Soluzione Esercizio 3:
def processa_immagine(img):
    risultati = {}
    
    # Grayscale: media pesata dei canali RGB
    # Pesi standard: 0.299*R + 0.587*G + 0.114*B
    pesi = np.array([0.299, 0.587, 0.114])
    grayscale = np.dot(img, pesi)
    risultati['grayscale'] = grayscale
    
    # Flip orizzontale
    flipped_h = np.flip(img, axis=1)
    risultati['flipped_h'] = flipped_h
    
    # Aumenta luminosità del 20%
    brightness = img * 1.2
    # Clip per mantenere valori in [0, 1]
    brightness = np.clip(brightness, 0, 1)
    risultati['brightness'] = brightness
    
    return risultati

# Test
img_test = np.random.rand(4, 4, 3)
risultati = processa_immagine(img_test)
print("Forme risultati:")
for k, v in risultati.items():
    print(f"{k}: {v.shape}")

### Soluzione Esercizio 4:
def kmeans_semplice(punti, k=3, iterazioni=10):
    n_punti = punti.shape[0]
    
    # Inizializza centroidi casuali
    indici_casuali = np.random.choice(n_punti, k, replace=False)
    centroidi = punti[indici_casuali].copy()
    
    for _ in range(iterazioni):
        # Calcola distanze da ogni punto a ogni centroide
        distanze = np.zeros((n_punti, k))
        for i in range(k):
            # Distanza euclidea vettorizzata
            diff = punti - centroidi[i]
            distanze[:, i] = np.sqrt(np.sum(diff**2, axis=1))
        
        # Assegna ogni punto al centroide più vicino
        assegnazioni = np.argmin(distanze, axis=1)
        
        # Aggiorna centroidi
        for i in range(k):
            punti_cluster = punti[assegnazioni == i]
            if len(punti_cluster) > 0:
                centroidi[i] = np.mean(punti_cluster, axis=0)
    
    return centroidi, assegnazioni

# Test
np.random.seed(42)
cluster1 = np.random.randn(30, 2) + [0, 0]
cluster2 = np.random.randn(30, 2) + [5, 5]
cluster3 = np.random.randn(30, 2) + [0, 5]
punti = np.vstack([cluster1, cluster2, cluster3])

centroidi, assegnazioni = kmeans_semplice(punti, k=3)
print(f"Centroidi trovati:\n{centroidi}")

# Verifica risultati
for i in range(3):
    n_punti = np.sum(assegnazioni == i)
    print(f"Cluster {i}: {n_punti} punti")
```
